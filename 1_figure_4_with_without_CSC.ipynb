```python
# treatments under warming and eCO2 - control
import pandas as pd
import numpy  as np

file_path_co2   = "/Users/jianzhou/MyDocs/1_works/1_proj_SPRUCE/1_SPRUCE_DA/4_summary_data_code_paper/test_output/2_mod_results/"
file_path_noacc = "/Users/jianzhou/MyDocs/1_works/1_proj_SPRUCE/1_SPRUCE_DA/4_summary_data_code_paper/test_output/3_mod_without_acc/"

var_names_in = ["plot", "warming",'year',
                "gpp_Tree",    "gpp_Shrub",    "gpp_Sphagnum", 
                "npp_Tree",    "npp_Shrub",    "npp_Sphagnum", 
                "ra_Tree",     "ra_Shrub",     "ra_Sphagnum",    
                "rh",          "ch4", "nce"]

var_names = ["gpp_Tree",    "gpp_Shrub",    "gpp_Sphagnum", 
                "npp_Tree",    "npp_Shrub",    "npp_Sphagnum", 
                "ra_Tree",     "ra_Shrub",     "ra_Sphagnum",    
                "rh",          "ch4", "nce"]

i_warmPlots  = [1, 9, 5, 2, 7]
i_warmEleCO2 = [8, 4, 0, 6, 3]
warming      = [0, 2.25, 4.5, 6.75, 9]

df_data_mlm_co2   = pd.read_excel(file_path_co2   + "results_MLM/1_2_1_data_MLM/2_df_res_mlm_yrDel_nomean.xlsx")
df_data_mlm_noacc = pd.read_excel(file_path_noacc + "results_MLM/1_2_1_data_MLM/2_df_res_mlm_yrDel_nomean.xlsx")


def cal_df_data(df_data_mlm):
    df_data = df_data_mlm[var_names_in]
    df_data = df_data[df_data["year"] > 2016]
    df_data["ra_Tree"]     = df_data["ra_Tree"]
    df_data["ra_Shrub"]    = df_data["ra_Shrub"]
    df_data["ra_Sphagnum"] = df_data["ra_Sphagnum"]
    df_data["rh"]          = df_data["rh"]
    df_data["ch4"]         = df_data["ch4"]
    return df_data

df_data_co2   = cal_df_data(df_data_mlm_co2)
df_data_noacc = cal_df_data(df_data_mlm_noacc)

# print(df_data_co2)

df_data_diff_co2   = pd.DataFrame(columns=var_names_in)

df_data = df_data_co2.copy()
for idx, (iwp, icp) in enumerate(zip(i_warmPlots, i_warmEleCO2)):
    print(iwp, icp)
    for year in df_data["year"].unique():
        dict_value = {}
        dict_value["plot"] = icp
        dict_value["year"] = year
        for ivar in var_names:
            diff = df_data.loc[(df_data['plot'] == icp) & (df_data['year'] == year), ivar].values[0] - df_data.loc[(df_data['plot'] == iwp) & (df_data['year'] == year), ivar].values[0]
            dict_value[ivar] = diff
            
        df_data_diff_co2 = pd.concat([df_data_diff_co2, pd.DataFrame([dict_value])], ignore_index=True)


df_data_diff_noacc = pd.DataFrame(columns=var_names_in)

for idx, (iwp, icp) in enumerate(zip(i_warmPlots, i_warmEleCO2)):
    print(iwp, icp)
    for year in df_data_noacc["year"].unique():
        dict_value = {}
        dict_value["plot"] = icp
        dict_value["year"] = year
        for ivar in var_names:
            diff = df_data_noacc.loc[(df_data_noacc['plot'] == icp) & (df_data_noacc['year'] == year), ivar].values[0] \
                 - df_data_noacc.loc[(df_data_noacc['plot'] == iwp) & (df_data_noacc['year'] == year), ivar].values[0]
            dict_value[ivar] = diff
        df_data_diff_noacc = pd.concat([df_data_diff_noacc, pd.DataFrame([dict_value])], ignore_index=True)


# print(df_data[["plot", 'warming', 'year', "gpp_Tree",    "gpp_Shrub",    "gpp_Sphagnum"]])
print(df_data_diff_co2)
print(df_data_diff_noacc)



import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np
from scipy import stats
# ----------------
# read data
# data with acclimation 
data_acc    = df_data_diff_co2[df_data_diff_co2['plot'].isin([8, 4, 0, 6, 3])]
data_nonacc = df_data_diff_noacc[df_data_diff_noacc['plot'].isin([8, 4, 0, 6, 3])]

colors = {"gpp_Shrub": "#337306", "gpp_Tree": "#4f942a",  "gpp_Sphagnum": "#60b333", 
          "ra_Shrub" : "#9ED1C8", "ra_Tree" : "#8AB0CF",  "ra_Sphagnum": "#BFBBD9",  
          "rh"       : "#F1B671",  "ch4"    : "#EB8776"}#,  "nce": "grey"}

ls_vars = ["gpp_Shrub", "gpp_Tree", "gpp_Sphagnum",
           "ra_Shrub",  "ra_Tree",  "ra_Sphagnum",
           "rh", "ch4"]
ls_plot = [8, 4, 0, 6, 3]

# color_nonacc = "#8ea0cb"
# color_acc    = "#fc8e62"

color_nonacc = "#f26350"#"#f0988c" # "#e88482" #"#fc8e62"# 
color_acc    = "#8e8bfe" #"#9394e7" #

import matplotlib.ticker as ticker
def to_int(x, pos):
    return '{:.0f}'.format(x)

import matplotlib.patches as patches
legend_handles = [ 
    patches.Rectangle((0,0),1,1, color = color_nonacc, linewidth=2, label='Fixed parameters'),#'Non-adjustment') 
    patches.Rectangle((0,0),1,1, color = color_acc, linewidth=2, label='Varying parameters'),#'Adjustment'),
]

# create a figure with 10 rows and 5  
# fig = plt.figure(figsize=(15, 7))
# gs  = gridspec.GridSpec(10, 6, figure=fig)

fig, axes = plt.subplots(3, 2,figsize=(12, 12))

bar_width = 0.35
x_acc = np.arange(8) + 0.5*bar_width
x_non = np.arange(8) - 0.5*bar_width

# each flux components: GPPshrub; GPPtree; GPPsphag.; Rashrub; Ratree; Rasphag.; Rh; CH4
ls_subtitle = ["+0 ℃ (a)", "+2.25 ℃ (b)", "+4.5 ℃ (c)", "+6.75 ℃ (d)", "+9 ℃ (e)"]
ls_sub   = ["(b)",    "(c)",     "(d)", "(e)",  "(f)"]
# ls_sub   = ["(b) +0 ℃",    "(c) +2.25 ℃",     "(d) +4.5 ℃", "(e) +6.75 ℃",  "(f) +9 ℃"]
ls_title = ["+0 ℃", "+2.25 ℃", "+4.5 ℃", "+6.75 ℃", "+9 ℃"]

ls_subplot = [axes[1,0], axes[2,0], axes[0,1], axes[1,1], axes[2,1]]


for i in range(5):
    # ax = fig.add_subplot(gs[i*2:(i+1)*2, 2:])
    ax = ls_subplot[i]
    ax.axhline(0, color='black', linestyle='-', linewidth=0.5, zorder=1)
    ax.text(0.99, 0.97, ls_title[i], fontsize=14, transform=ax.transAxes, ha='right', va='top')
    ax.text(0.015, 0.97, ls_sub[i], fontsize=14, transform=ax.transAxes, ha='left', va='top')
    # if i == 4:
    #     ax.text(0.01, 0.08, ls_subtitle[i], fontsize=14, transform=ax.transAxes)
    # else:
    #     ax.text(0.01, 0.85, ls_subtitle[i], fontsize=14, transform=ax.transAxes)
    for idx, ivar in enumerate(ls_vars): # each variable 
        # plot non adjustment
        plt_x = np.full(5, x_non[idx])
        plt_y = data_nonacc[data_nonacc["plot"]==ls_plot[i]][ivar]
        plt_y_noacc = plt_y.to_numpy().flatten()
        if i == 4 and idx == 0: gpp_shrub_non = plt_y
        if i == 4 and idx == 6: rh_shrub_non  = plt_y
        # ax.scatter(plt_x, plt_y, color="#2F81B5",)
        # ax.errorbar(x_non[idx], np.mean(plt_y), yerr=[np.array(np.std(plt_y)), np.array(np.std(plt_y))])
        if ivar == "gpp_Tree": print("test_value_tree: ", x_non[idx], np.mean(plt_y), np.std(plt_y))
        if ivar == "gpp_Sphagnum": print("test_value_tree: ", x_non[idx], np.mean(plt_y), np.std(plt_y))
        ax.bar(x_non[idx], np.mean(plt_y), yerr=np.std(plt_y), width=bar_width, capsize=2, color=color_nonacc, alpha=1, error_kw=dict(ecolor="black", lw=0.7, capthick=0.7), edgecolor="black", linewidth=0.7) #linewidth=2, facecolor='none')
        # plot adjustment
        plt_x = np.full(5, x_acc[idx])
        plt_y = data_acc[data_acc["plot"]==ls_plot[i]][ivar]
        plt_y_acc = plt_y.to_numpy().flatten()
        # print(plt_y_noacc, plt_y_acc)
        # t_stat, p_value = stats.ttest_ind(plt_y_noacc, plt_y_acc)
        t_stat, p_value = stats.ttest_rel(plt_y_noacc, plt_y_acc)
        # print("statistic: ", idx, "--", t_stat, p_value)
        if p_value < 0.001: mark_p_0 = "***"
        elif p_value < 0.01: mark_p_0 = "**"
        elif p_value < 0.05: mark_p_0 = "*"
        else: mark_p_0 =""
        ax.text(idx, -380, mark_p_0, ha="center", va="center", fontsize=10)
        if i == 4 and idx == 0: 
            gpp_shrub_acc = plt_y
            print("gpp_shrub: ", np.mean(gpp_shrub_acc - gpp_shrub_non), np.std(gpp_shrub_acc - gpp_shrub_non))
        if i == 4 and idx == 6:
            rh_shrub_acc = plt_y
            print("rh_shrub: ", np.mean(rh_shrub_acc - rh_shrub_non), np.std(rh_shrub_acc - rh_shrub_non))

        # ax.scatter(plt_x, plt_y, color='#E51818')
        ax.bar(x_acc[idx], np.mean(plt_y), yerr=np.std(plt_y), width=bar_width, capsize=2, color=color_acc, alpha=1, error_kw=dict(ecolor="black", lw=0.7, capthick=0.7), edgecolor="black", linewidth=0.7) #, linewidth=2, facecolor='none')
    ax.set_ylim(-400, 800)

    # if i == 2: ax.text(-0.1, 0.5, "CO$_2$ effect (g C m$^{-2}$ yr$^{-1}$)", ha='center', va='center',rotation='vertical',transform=ax.transAxes, fontsize=16)
    ax.set_yticklabels(ax.get_yticks(), fontsize=12)
    ax.yaxis.set_major_formatter(ticker.FuncFormatter(to_int))

    # if i == 2: ax.legend(bbox_to_anchor=(0.62, 0.99), handles=legend_handles)
    if i == 0: ax.legend("upper center", bbox_to_anchor=(0.78, 0.99),handles=legend_handles)

    ax.set_yticks([-300, 0, 300, 600], [-300, 0, 300, 600], fontsize=12) 
    ax.set_xticklabels([])

    custom_x_ticks = np.arange(8)
    custom_x_labels = ["GPP$_{Shrub}$", "GPP$_{Tree}$", "GPP$_{Sphag.}$", "Ra$_{Shrub}$", "Ra$_{Tree}$", "Ra$_{Sphag.}$", "Rh$_{CO_2}$", "CH$_4$"]
    ax.set_xticks(custom_x_ticks, custom_x_labels,rotation=45, fontsize=14)  

# axes[2,0].set_xticks(custom_x_ticks, custom_x_labels, fontsize=11)
# ax.text(0.5, -0.35, "Components", fontsize=16, ha='center', va='center', transform=ax.transAxes)

# plot the NCE
import seaborn as sns
from scipy.stats import linregress
ls_warming = [0, 2.25, 4.5, 6.75, 9]

i = 0

ax = axes[0,0]


ax.text(0.015, 0.97, "(a) NCE", fontsize=14, transform=ax.transAxes, ha='left', va='top')
ls_x = []
ls_y = []
for idx, iwarm in enumerate(ls_plot):
    plt_x = np.full(5, ls_warming[idx])
    plt_y = data_nonacc[data_nonacc["plot"]==iwarm]["nce"]
    t_stat, p_value = stats.ttest_1samp(plt_y, 0)
    # print("statistic: ", iwarm, "--", t_stat, p_value)
    # ax.scatter(plt_x, plt_y)
    ls_x.extend(plt_x)
    ls_y.extend(plt_y)
    ax.errorbar(ls_warming[idx], np.mean(plt_y), yerr=np.std(plt_y), fmt='o', ecolor=color_nonacc, capsize=3, mfc='none', mec=color_nonacc, mew=1.5)
    if p_value < 0.001: mark_p_0 = "***"
    elif p_value < 0.01: mark_p_0 = "**"
    elif p_value < 0.05: mark_p_0 = "*"
    else: mark_p_0 =""
    ax.text(ls_warming[idx], -110, mark_p_0, ha="center", va="center", fontsize=10)
ax.set_yticklabels(ax.get_yticks(), fontsize=12)
ax.yaxis.set_major_formatter(ticker.FuncFormatter(to_int))

slope, intercept, r_value, p_value, std_err = linregress(ls_x, ls_y)
if p_value < 0.001: mark_p = "***"; line_style = "-"
elif p_value < 0.01: mark_p = "**"; line_style = "-"
elif p_value < 0.05: mark_p = "*"; line_style = "-"
else: mark_p = ""; line_style = "--"

sns.regplot(x=ls_x, y=ls_y, ax=ax, scatter_kws={'alpha': 0.5, 'color':color_nonacc},line_kws={"linestyle": line_style,'linewidth': 3},  color=color_nonacc) #, color="#2F81B5",  label="Ambient CO$_2$")
ax.text(0.02, 0.78, "y = {:.1f}x + {:.1f}; R$^2$ = {:.2f}".format(slope,intercept,r_value*r_value)+"$^{"+mark_p+"}$", fontsize=14, transform=ax.transAxes, color=color_nonacc, va='bottom')
axes[1,0].text(-0.15, 0.5, "CO$_2$ effect (g C m$^{-2}$ yr$^{-1}$)", ha='center', va='center',rotation='vertical',transform=axes[1,0].transAxes, fontsize=16)
# 
ax.set_xticks([0, 2.25, 4.5, 6.75, 9], [], fontsize=12)
# ax.set_ylim(-299, 499)


ax.set_xticks([0, 2.25, 4.5, 6.75, 9], ["+0 ℃", "+2.25 ℃", "+4.5 ℃", "+6.75 ℃", "+9 ℃"], fontsize=12)
# ax.text(0.5, -0.13, "Warming treatment (℃)", fontsize=16, ha='center', va='center', transform=ax.transAxes)
ax.set_ylim(-119, 399)

plt.subplots_adjust(wspace=0.15, hspace=0.35)
```
